```python
import pandas as pd
```

```python
df=pd.read_csv("twitterja.csv",index_col="時間") #CSVファイルを設定します。
```

```python
df
```

```python
new_column_order = ['ツイート本文','インプレッション','エンゲージメント','エンゲージメント率','ユーザープロフィールクリック','リツイート','いいね','返信','詳細クリック','メディアの再生数','メディアのエンゲージメント数'] 
df = df[new_column_order]
```

```python
df
```

```python
target_columns = ['インプレッション', 'エンゲージメント', 'ユーザープロフィールクリック','リツイート','いいね','返信','詳細クリック']
df[target_columns] = df[target_columns].astype(int)
```

```python
dtarget_column = "エンゲージメント率"
df[dtarget_column] = df[dtarget_column] * 100
decimal_places = 0
df[dtarget_column] = df[dtarget_column].round(decimal_places)
df[dtarget_column] = df[dtarget_column].astype(str) + '%'
```

```python
df
```

```python
df.to_excel('x.xlsx') #Excelへ保存
```
